```python
a = df['Input table']['Inputs for logical operations']['Hex2Bin']
b = df['Input table']['Inputs for logical operations']['Hex2Bin.1']
a_bin=[]
b_bin=[]
for i in a:
    binary=bin(int(i, 16))[2:]
    a_bin.append(binary)
    
for i in b:
    binary=bin(int(i, 16))[2:]
    b_bin.append(binary)

b_bin = [n.zfill(16) for n in b_bin]

A = pd.Series(a_bin)
B = pd.Series(b_bin)
```

```python
#logical_and
result=[]
for i in range(len(a_bin)):
    and_result=''
    for j in range(len(a_bin[i])):
        and_result += str(int(a_bin[i][j]) & int(b_bin[i][j]))
    result.append(and_result)
a = []
for i in result:
    groups = [i[j:j+4] for j in range(0, len(i), 4)]
    hex_digits=''
    for group in groups:
        hex_digits += hex(int(group, 2))[2:]
    ans = hex_digits[3:]
    a.append(ans)
AND = pd.Series(a)
df.loc[:, ('Output table', 'Binary Operation to be printed in subsequent foldera', 'AND')]= AND
```

```python
#logical_OR
result=[]
for i in range(len(a_bin)):
    or_result=''
    for j in range(len(a_bin[i])):
        or_result += str(int(a_bin[i][j]) | int(b_bin[i][j]))
    result.append(or_result)
a = []
for i in result:
    groups = [i[j:j+4] for j in range(0, len(i), 4)]
    hex_digits=''
    for group in groups:
        hex_digits += hex(int(group, 2))[2:]
    ans = hex_digits
    a.append(ans)
OR = pd.Series(a)
df.loc[:, ('Output table', 'Binary Operation to be printed in subsequent foldera', 'OR')]= OR
```

```python
#logical_NOR
l=[]
for i in OR:
    nor_val = hex(~int(i, 16) & 0xFFFF)[2:]
    l.append(nor_val)
NOR = pd.Series(l)
df.loc[:, ('Output table', 'Binary Operation to be printed in subsequent foldera', 'NOR')]= NOR
```

```python
#logical_NAND
l=[]
for i in AND:
    nand_val = hex(~int(i, 16) & 0xFFFF)[2:]
    l.append(nand_val)
NAND = pd.Series(l)
df.loc[:, ('Output table', 'Binary Operation to be printed in subsequent foldera', 'NAND')]= NAND
```

```python
#logical_XOR
result=[]
for i in range(len(a_bin)):
    xor_result=''
    for j in range(len(a_bin[i])):
        xor_result += str(int(a_bin[i][j]) ^ int(b_bin[i][j]))
    result.append(xor_result)
a = []
for i in result:
    groups = [i[j:j+4] for j in range(0, len(i), 4)]
    hex_digits=''
    for group in groups:
        hex_digits += hex(int(group, 2))[2:]
    ans = hex_digits
    a.append(ans)
XOR = pd.Series(a)
df.loc[:, ('Output table', 'Binary Operation to be printed in subsequent foldera', 'XOR')]= XOR
```

```python
#logical_XNOR
l=[]
for i in XOR:
    nand_val = hex(~int(i, 16) & 0xFFFF)[2:]
    l.append(nand_val)
XNOR = pd.Series(l)
df.loc[:, ('Output table', 'Binary Operation to be printed in subsequent foldera', 'XNOR')]= XNOR
```
